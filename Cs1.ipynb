### 3. Perform EDA
#### Assignment: 
1. Perform EDA on cleaned data.(Hint plot the data) 
2. Write atleat two observation from EDA.

```python
import matplotlib.pyplot as plt

# Cleaned data
studied_hours = [25, 7, 10, 15]
grade = [10, 7, 8, 10]

# Scatter plot
plt.scatter(studied_hours, grade, color='blue')
plt.title("Studied Hours vs Grade")
plt.xlabel("Studied Hours")
plt.ylabel("Grade")
plt.grid(True)
plt.show()
```

### 📊 Observations from EDA

1. **Positive Relationship Between Studied Hours and Grade:**  
   Students who studied more generally scored higher grades.  
   **Example:** Naman studied **25 hours** and scored **10**, while Ankit studied only **7 hours** and scored **7**.

2. **Diminishing Returns After a Point:**  
   After around **15 hours**, increasing study time doesn't seem to raise grades further.  
   **Example:** Ramu studied **15 hours** and scored **10**, the same as Naman who studied **25 hours**.


### 4. Apply Techniques
#### Assignment: 
1. Apply a machine learning or data science technique to encode your observation in EDA. (Hint: which type of realtion is associated between hours of study and grade)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
studied_hours = np.array([25, 7, 10, 15]).reshape(-1, 1)
grades = np.array([10, 7, 8, 10])

# Model
model = LinearRegression()
model.fit(studied_hours, grades)

# Predict
predicted = model.predict(studied_hours)

# Plot
plt.scatter(studied_hours, grades, color='blue', label='Actual Data')
plt.plot(studied_hours, predicted, color='red', label='Regression Line')
plt.title("Simple Linear Regression: Studied Hours vs Grade")
plt.xlabel("Studied Hours")
plt.ylabel("Grade")
plt.legend()
plt.grid(True)
plt.show()

# Display slope and intercept
print(f"Model: Grade = {model.coef_[0]:.2f} * Hours + {model.intercept_:.2f}")
```
